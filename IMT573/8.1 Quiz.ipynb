The Boston dataset records `medv` (median house value) for neighborhoods around Boston.<p>
Predict `medv` using predictors such as `rmvar` (average number of rooms per house), age (proportion of owner-occupied units built prior to 1940), and `lstat` (percent of households with low socioeconomic status).

```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('Boston.csv')
display(df)
```

```python
X = sm.add_constant(df['lstat'])

# Define the response variable (runs)
y = df['medv']

# Fit the linear regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```
